```python
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dropout, Conv1D, GlobalMaxPooling1D, Dense
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import ParameterGrid
from tensorflow.keras.callbacks import EarlyStopping
from itertools import product

import numpy as np
import matplotlib.pyplot as plt
```

# Download Dataset

```python
vocab_size = 5000 
maximum_sequence_length = 500

(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words= vocab_size)

x_train, x_val = x_train[:-5000], x_train[-5000:]
y_train, y_val = y_train[:-5000], y_train[-5000:]

INDEX_FROM = 3
word_index = imdb.get_word_index()
word_index = {key:(value+INDEX_FROM) for key,value in word_index.items()}
word_index["<PAD>"] = 0
word_index["<START>"] = 1
word_index["<UNK>"] = 2
reverse_word_index = {value:key for key, value in word_index.items()}


x_train = pad_sequences(x_train, value= word_index["<PAD>"], padding= 'post', maxlen= maximum_sequence_length)
x_val = pad_sequences(x_val, value= word_index["<PAD>"], padding= 'post', maxlen= maximum_sequence_length)
x_test = pad_sequences(x_test, value= word_index["<PAD>"], padding= 'post', maxlen= maximum_sequence_length)
```

# Model


```python
embedding_dim = 16

def create_model(vocab_size, maximum_sequence_length, filters=64, kernel_size=3, strides=1, 
                 units=256, optimizer='adam', rate=0.5, kernel_initializer='glorot_uniform'):
    model = Sequential()
    model.add(Embedding(vocab_size, embedding_dim, input_shape=(maximum_sequence_length,)))
    model.add(Dropout(rate))
    model.add(Conv1D(filters=filters, kernel_size=kernel_size, strides=strides,
                     padding='same', activation='relu'))
    model.add(GlobalMaxPooling1D())
    model.add(Dense(units=units, activation='relu', kernel_initializer=kernel_initializer))
    model.add(Dropout(rate))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

# параметры для перебора
filters_list = [128]
kernel_size_list = [5]
strides_list = [1]
units_list = [128, 512]
kernel_initializer_list = ['glorot_uniform']
rate_list = [0.25]
optimizer_list = ['adam']
epochs_list = [5]
batch_list = [64]

param_grid = list(product(filters_list, kernel_size_list, strides_list, units_list,
                          kernel_initializer_list, rate_list, optimizer_list,
                          epochs_list, batch_list))

best_score = 0
best_params = None
```

```python
for params in param_grid:
    f, k, s, u, ki, r, opt, ep, bs = params
    print("Testing parameters:", params)
    model = create_model(vocab_size, maximum_sequence_length,
                         filters=f, kernel_size=k, strides=s, units=u,
                         kernel_initializer=ki, rate=r, optimizer=opt)
    
    earlystopper = EarlyStopping(monitor='val_accuracy', patience=0, verbose=1)
    
    history = model.fit(x_train, y_train,
                        validation_data=(x_val, y_val),
                        epochs=ep,
                        batch_size=bs,
                        callbacks=[earlystopper],
                        shuffle=True,
                        verbose=0)
    
    score = history.history['val_accuracy'][-1]
    if score > best_score:
        best_score = score
        best_params = params

print("Best val_accuracy:", best_score)
print("Best parameters:", best_params)
```

# Learning Model

```python
f, k, s, u, ki, r, opt, ep, bs = best_params

final_model = create_model(vocab_size, maximum_sequence_length,
                           filters=f, kernel_size=k, strides=s, units=u,
                           kernel_initializer=ki, rate=r, optimizer=opt)

history = final_model.fit(
    np.vstack((x_train, x_val)),
    np.hstack((y_train, y_val)),
    validation_split=0.2,
    epochs=ep,
    batch_size=bs,
    shuffle=True,
    verbose=1
)
```

## Training and validation loss

```python
history_dict = history.history
epochs = range(1, len(history_dict['loss'])+1)
plt.plot(epochs, history_dict['loss'], 'bo', label='Training Loss')
plt.plot(epochs, history_dict['val_loss'], 'b', label='Validation Loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

## Training and validation accuracy

```python
plt.plot(epochs, history_dict['accuracy'], 'bo', label='Training Accuracy')
plt.plot(epochs, history_dict['val_accuracy'], 'b', label='Validation Accuracy')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

# Test Accuracy

```python
test_loss, test_accuracy = final_model.evaluate(x_test, y_test)
```
