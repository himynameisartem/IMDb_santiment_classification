```python
# Импорты для работы с данными и моделью
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SpatialDropout1D, Bidirectional, LSTM, GlobalMaxPooling1D, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

import numpy as np
import matplotlib.pyplot as plt
import os
import zipfile
import urllib.request
```

# Download data

### Downloading IMDb

```python
vocab_size = 20000  # Увеличиваем словарь для лучшего качества
maxlen = 250  # Оптимальная длина последовательности

print("Downloading IMDB...")
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=vocab_size)

# Разделяем train на train и validation (5000 примеров для валидации)
x_train, x_val = x_train[:-5000], x_train[-5000:]
y_train, y_val = y_train[:-5000], y_train[-5000:]

# Приводим последовательности к одной длине (pad_sequences)
x_train = pad_sequences(x_train, maxlen=maxlen, padding='post')
x_val = pad_sequences(x_val, maxlen=maxlen, padding='post')
x_test = pad_sequences(x_test, maxlen=maxlen, padding='post')

print(f"Data size:")
print(f"Train: {x_train.shape}, Val: {x_val.shape}, Test: {x_test.shape}")
```

### Downloading Glove Embeddings

```python
# GloVe - это предобученные векторы слов (word embeddings)
# Они помогают модели понимать семантику слов лучше, чем случайная инициализация

print("Downloading GloVe embeddings...")

# Скачиваем GloVe 100d (если еще не скачан)
glove_file = 'glove.6B.100d.txt'
glove_url = 'https://nlp.stanford.edu/data/glove.6B.zip'

if not os.path.exists(glove_file):
    print("Скачивание GloVe (это может занять несколько минут)...")
    zip_file = 'glove.6B.zip'
    
    if not os.path.exists(zip_file):
        urllib.request.urlretrieve(glove_url, zip_file)
        print("Архив скачан")
    
    # Распаковываем только нужный файл
    with zipfile.ZipFile(zip_file, 'r') as zip_ref:
        zip_ref.extract(glove_file)
    print("GloVe распакован")
else:
    print("GloVe уже загружен")

# Загружаем GloVe векторы в словарь
print("Чтение GloVe векторов...")
embeddings_index = {}
with open(glove_file, encoding='utf-8') as f:
    for line in f:
        values = line.split()
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        embeddings_index[word] = coefs

print(f"Загружено {len(embeddings_index)} векторов слов")
```

# Embedding

## Create embedding matrix

```python
# ============================================
# ШАГ 3: СОЗДАНИЕ EMBEDDING МАТРИЦЫ
# ============================================
# Нужно сопоставить слова из IMDB с векторами GloVe
# Создаем матрицу размером (vocab_size, embedding_dim)

print("Создание embedding матрицы...")

# Получаем словарь IMDB
INDEX_FROM = 3
word_index = imdb.get_word_index()
word_index = {k: (v + INDEX_FROM) for k, v in word_index.items()}
word_index["<PAD>"] = 0
word_index["<START>"] = 1
word_index["<UNK>"] = 2

# Размерность embedding (GloVe 100d)
embedding_dim = 100

# Создаем матрицу: сначала заполняем случайными значениями
embedding_matrix = np.random.normal(scale=0.6, size=(vocab_size, embedding_dim)).astype('float32')

# Заполняем матрицу векторами GloVe где возможно
found = 0
for word, i in word_index.items():
    if i < vocab_size:
        embedding_vector = embeddings_index.get(word)
        if embedding_vector is not None:
            embedding_matrix[i] = embedding_vector
            found += 1

print(f"Найдено совпадений: {found} из {vocab_size}")
print(f"Покрытие: {found/vocab_size*100:.1f}%")
```

# Model

## Create model

```python
# ============================================
# ШАГ 4: СОЗДАНИЕ МОДЕЛИ
# ============================================
# Архитектура модели:
# 1. Embedding слой - преобразует индексы слов в векторы (используем GloVe)
# 2. SpatialDropout1D - регуляризация для embedding
# 3. Bidirectional LSTM - двунаправленная LSTM для понимания контекста
# 4. GlobalMaxPooling1D - берем максимальные значения по последовательности
# 5. Dense(128) - полносвязный слой
# 6. Dropout - регуляризация
# 7. Dense(1, sigmoid) - выходной слой для бинарной классификации

def create_model(trainable_embedding=False):
    """
    Создает модель для классификации тональности
    
    Параметры:
    trainable_embedding: если False - embedding слой ЗАМОРОЖЕН (веса не меняются)
                         если True - embedding слой РАЗМОРОЖЕН (веса обучаются)
    
    Что такое замораживание?
    - trainable=False: используем предобученные GloVe векторы как есть
    - trainable=True: позволяем модели подстроить векторы под нашу задачу
    """
    model = Sequential([
        # Embedding слой: преобразует индексы слов в векторы
        Embedding(
            input_dim=vocab_size,           # размер словаря
            output_dim=embedding_dim,       # размерность вектора (100 для GloVe)
            weights=[embedding_matrix],     # используем предобученные GloVe векторы
            input_length=maxlen,            # длина входной последовательности
            trainable=trainable_embedding   # ЗАМОРАЖИВАЕМ или РАЗМОРАЖИВАЕМ слой
        ),
        
        # SpatialDropout1D - удаляет целые векторы слов (лучше чем обычный Dropout для embedding)
        SpatialDropout1D(0.2),
        
        # Bidirectional LSTM - двунаправленная LSTM
        # LSTM читает последовательность в обоих направлениях (слева направо и справа налево)
        # return_sequences=True - возвращаем выход для каждого слова в последовательности
        Bidirectional(LSTM(64, return_sequences=True, recurrent_dropout=0.2)),
        
        # GlobalMaxPooling1D - берем максимальное значение по каждому измерению
        # Это уменьшает размерность и выделяет самые важные признаки
        GlobalMaxPooling1D(),
        
        # Полносвязный слой с активацией ReLU
        Dense(128, activation='relu'),
        
        # Dropout для регуляризации (предотвращает переобучение)
        Dropout(0.5),
        
        # Выходной слой: 1 нейрон с сигмоидой (вероятность положительного отзыва)
        Dense(1, activation='sigmoid')
    ])
    
    # Компилируем модель
    model.compile(
        loss='binary_crossentropy',  # функция потерь для бинарной классификации
        optimizer=Adam(learning_rate=2e-3),  # оптимизатор с learning rate
        metrics=['accuracy']  # метрика для оценки
    )
    
    return model

# Выводим архитектуру модели
print("Архитектура модели:")
model = create_model(trainable_embedding=False)
model.summary()
```

## Learning Model

```python
# ============================================
# ШАГ 5: ОБУЧЕНИЕ МОДЕЛИ (ДВУХЭТАПНОЕ)
# ============================================

# Callbacks (обратные вызовы):
# - EarlyStopping: останавливает обучение если нет улучшения
# - ReduceLROnPlateau: уменьшает learning rate если модель не улучшается

early_stopping = EarlyStopping(
    monitor='val_accuracy',      # следим за точностью на валидации
    patience=2,                  # ждем 2 эпохи без улучшения
    restore_best_weights=True,   # возвращаем лучшие веса
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',          # следим за loss на валидации
    factor=0.5,                  # уменьшаем LR в 2 раза
    patience=1,                  # ждем 1 эпоху
    verbose=1
)

# ============================================
# ЭТАП 1: ОБУЧЕНИЕ С ЗАМОРОЖЕННЫМ EMBEDDING
# ============================================
# Сначала обучаем модель с замороженными GloVe векторами
# Это позволяет модели научиться использовать предобученные векторы

print("\n" + "="*60)
print("ЭТАП 1: Обучение с ЗАМОРОЖЕННЫМ embedding слоем")
print("="*60)

model = create_model(trainable_embedding=False)  # ЗАМОРАЖИВАЕМ embedding

history1 = model.fit(
    x_train, y_train,
    validation_data=(x_val, y_val),
    epochs=3,                    # обучаем 3 эпохи
    batch_size=64,
    callbacks=[early_stopping, reduce_lr],
    shuffle=True,
    verbose=1
)

print(f"\nЛучшая точность на валидации (этап 1): {max(history1.history['val_accuracy']):.4f}")

# ============================================
# ЭТАП 2: ДООБУЧЕНИЕ С РАЗМОРОЖЕННЫМ EMBEDDING
# ============================================
# Теперь размораживаем embedding и дообучаем модель
# Это позволяет подстроить векторы слов под нашу конкретную задачу

print("\n" + "="*60)
print("ЭТАП 2: Дообучение с РАЗМОРОЖЕННЫМ embedding слоем")
print("="*60)

# РАЗМОРАЖИВАЕМ embedding слой
model.layers[0].trainable = True

# Перекомпилируем с меньшим learning rate (для более аккуратной донастройки)
model.compile(
    loss='binary_crossentropy',
    optimizer=Adam(learning_rate=1e-3),  # меньший LR для тонкой настройки
    metrics=['accuracy']
)

# Дообучаем модель
history2 = model.fit(
    x_train, y_train,
    validation_data=(x_val, y_val),
    epochs=8,                    # дообучаем до 8 эпох (early stopping остановит раньше)
    batch_size=64,
    callbacks=[early_stopping, reduce_lr],
    shuffle=True,
    verbose=1
)

# Объединяем историю обучения
history = {
    'loss': history1.history['loss'] + history2.history['loss'],
    'val_loss': history1.history['val_loss'] + history2.history['val_loss'],
    'accuracy': history1.history['accuracy'] + history2.history['accuracy'],
    'val_accuracy': history1.history['val_accuracy'] + history2.history['val_accuracy']
}

print(f"\nЛучшая точность на валидации (финальная): {max(history['val_accuracy']):.4f}")
```

# Charts

## Training and validation loss

```python

epochs = range(1, len(history['loss'])+1)
plt.plot(epochs, history['loss'], 'bo', label='Training Loss')
plt.plot(epochs, history['val_loss'], 'b', label='Validation Loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

## Training and validation accuracy

```python
plt.plot(epochs, history['accuracy'], 'bo', label='Training Accuracy')
plt.plot(epochs, history['val_accuracy'], 'b', label='Validation Accuracy')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

# Test Accuracy

```python
# ============================================
# ШАГ 6: ОЦЕНКА МОДЕЛИ НА ТЕСТОВОЙ ВЫБОРКЕ
# ============================================

print("\n" + "="*60)
print("ОЦЕНКА МОДЕЛИ")
print("="*60)

# Оценка на валидационной выборке
val_loss, val_accuracy = model.evaluate(x_val, y_val, verbose=0)
print(f"\nТочность на ВАЛИДАЦИОННОЙ выборке: {val_accuracy:.4f} ({val_accuracy*100:.2f}%)")

# Оценка на тестовой выборке
test_loss, test_accuracy = model.evaluate(x_test, y_test, verbose=0)
print(f"Точность на ТЕСТОВОЙ выборке: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")

print("\n" + "="*60)
if val_accuracy >= 0.95:
    print("✓ Цель достигнута на валидации (≥95%)")
else:
    print(f"⚠ На валидации: {val_accuracy*100:.2f}% (цель: 95%)")

if test_accuracy >= 0.93:
    print("✓ Цель достигнута на тесте (≥93%)")
else:
    print(f"⚠ На тесте: {test_accuracy*100:.2f}% (цель: 93%)")
print("="*60)
```
